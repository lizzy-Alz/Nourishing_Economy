```python
#supuestamente debe sacra una lista con todos los productos, otra con 
# los productos feemninos y otra con los masculinos de cada una de las mipymes
def  get_products_and_classifications(mipyme):
    list_prod=[]
    price_p=[]
    list_basic = []
    price_basic=[]
    products_m= []
    price_m=[]
    products_h= []
    price_h =[]
    
            
    for pro in mipyme["products"]:
        name = pro["name"]
        price = pro["price"]
        category = pro["category"]
        
        if name and price is not None and category:
            list_prod.append(name)
            price_p.append(price)
        
            if category == "íntima" or category == "protectores" or category == "tampones" or category == "crema de peinar" or category == "crema corporal" or category == "tinte" or category == "gel íntimo" or category == "protector solar":
                products_m.append(name)
                price_m.append(price)
                
            if category == "cuchillas de afeitar" or category == "after shave" or category == "crema de afeitar":
                products_h.append(name)
                price_h.append(price)
            else:
                list_basic.append(name)
                price_basic.append(price)
    
    return list_prod ,price_p,list_basic,price_basic, products_m, price_m, products_h,price_h
```
